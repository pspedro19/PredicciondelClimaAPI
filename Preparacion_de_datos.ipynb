# **Entendimiento del Negocio**

Durante la pandemia de SARS COVID-19 millones de personas alrededor del mundo se vieron afectadas por este virus, directa e indirectamente, desde la perspectiva de salud publica, millones perdieron su vida, mientras que algunos un tanto afortunados quedaron con secuelas como perdida del olfato o gusto. desde el punto de vista social-economico muchas otras perdieron sus empleoss, razon por la cual entidades gubernametales se emepzaron a cuestionar sus politicas de gobierno y como deberian tomar mejores decisiones apartir de una nueva situacion a nivel mundial, para impulsar el desarrollo economico.

Para comprender de mejor manera, el efecto de estas politicas, es importante observar cómo les va a ciertos países dentro de sus propias regiones, que han tenido historias similares de enfermedades y virus.

Los gobiernos de todo el mundo han implementado una serie de medidas diferentes en respuesta a los brotes de COVID-19. Estas incluyen medidas que tienen como objetivo contener la propagación del virus (como restricciones de movimiento y cierre de fronteras), medidas para mitigar las consecuencias económicas (como apoyo a los ingresos), así como medidas relacionadas con el sistema de salud (como políticas de prueba y rastreo de contactos). Miles de millones de personas se ven afectadas por estas medidas. Aunque los niños se han librado en gran medida de los efectos directos en la salud del COVID-19, los impactos indirectos, incluidos los enormes desafíos socioeconómicos, son potencialmente catastróficos para los niños. Para los niños migrantes y desplazados, los efectos pueden ser aún más graves.

A raiz de esto nos surge distintas incognitas y situaciones que quisieramos comprender más sobre el comportamiento de los paises y posible tendencias durante la pandemia, para asi crear politicas gubernamentales que puedan mejorar la calidad de vida de los ciudadanoss. Por tal razon es de gran interés conocer los paises que más fallecidos presentaron durante la pandemia, qué caracteristicas presentaban los paises más afectados y que relacion tenian con los menos afectados.

Adicional es importante ver la tendencia con la que se han presentado estos paises durante ese periodo de tiempo y asi remplatear las medidas actuales y proponer nuevas de ser necesarios

**4. Limpieza de datos**

```python
#Se asigna los valores nulos a los datos atipicos segun la regla de calidad
#Incapcidades deben estar en un rago entre o y 365 dias (Incapcidades en un año)

#no pueden haber dias negativos
datos.tiempo_inicioSintomas_diagnostico[datos["tiempo_inicioSintomas_diagnostico"]<0]=np.nan
datos.info()
```

```python
#considerar que la edad maxima por un humano en 100 años

datos.Edad[datos["Edad"]>100]=np.nan
datos.info()
```

**5. Limpieza de datos Nulos**

```python
#Limpieza de datos Nulos; Imputacion por media y moda

from sklearn.impute import SimpleImputer

#Imputacion de variables numericas
ImpNumeros=SimpleImputer(missing_values=np.nan, strategy='mean')
datos[['tiempo_inicioSintomas_diagnostico', 'Edad']] = ImpNumeros.fit_transform(datos[['tiempo_inicioSintomas_diagnostico', 'Edad']])

#Imputacion de variables categoricas

#ImpCategorias = SimpleImputer(missing_values=np.nan, strategy='most_frequent')
#datos[['Carro']]=ImpCategorias.fit_transform(datos[['Carro']])
datos.info()
```

```python
#Valores de la imputacion
print(ImpNumeros.statistics_)
#print(ImpCategorias.statistics_)
```

**7. Transformaciones segun el metodo de ML**

```python
#Sklearn solo trabaja con variables numericas
```

```python
#Normalizacion de las variables númericas

from sklearn.preprocessing import MinMaxScaler

min_max_scaler=MinMaxScaler()
min_max_scaler.fit(datos[['tiempo_inicioSintomas_diagnostico', 'Edad']])
datos[['tiempo_inicioSintomas_diagnostico', 'Edad']]=min_max_scaler.transform(datos[['tiempo_inicioSintomas_diagnostico', 'Edad']])
datos.head()
```
